```python
null_index_list = []
index = 0
eksik_veri = pd.isna(df['daily_vaccinations'])
for satir in eksik_veri:
    if satir == True:
        null_index_list.append(index)
    index+=1
    
    
```

```python
df = df.fillna(0)  ## medyanı daha doğru bulmak ve kuveyt gibi medyanı olmayan ülkeleri 0 yapmak için
```

```python
medyan_list = df.groupby('country').median()
medyan_list
```

```python
df.iloc[0]['country']
```

```python
ulkem = 'Turkey'
medyan_list.loc[ulkem].values[0]
```

```python
for nulls in null_index_list:
    ulke = df.iloc[nulls]['country']
    ulke_medyan = medyan_list.loc[ulke].values[0]
    df['daily_vaccinations'].iloc[nulls] = ulke_medyan
    
```

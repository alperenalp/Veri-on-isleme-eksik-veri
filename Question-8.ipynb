```python
siteler = []
for url in url_list:
    if url.find('https') == -1 :  ## string de https yoksa 
        site = url[12:-6]
        siteler.append(site)
    else:
        site = url[13:-6]
        siteler.append(site)
siteler
```

```python
lower_siteler = []
for i in range(len(siteler)):
    lower_siteler.append(siteler[i].lower())
print(lower_siteler)
```

```python
for index in df.index:
    df['Stats_Access_Link'].iloc[index]=lower_siteler[index]
```
